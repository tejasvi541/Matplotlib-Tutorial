```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Using subplots to plot the data we previously plotted using plt method. The plot below is the first plot we created in line plots

# We are using the plt objcet that we have imported we never actually created that object ourselves, 

# Figure - Figures are the containers holding our plots.
# Axes - these are the actual plots, so a figure may have many plots.
#  We have seen just one figure and one plot but we can also use many plots in one figure
# Working with many plots we can use subplot method

plt.style.use('seaborn')

data = pd.read_csv('data/SubPlotsData.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

# We will instantiate a figure and axis and then our axis is set to a single(1,1) axis but you can set it to rows and columns

# So this is useful when insted of having all of data in one plot we can divide them into multiple plots, so we can add more
# axis and figure

fig1, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex= True) # One figure and 2 axes, passing sharex = True so that both the plots
# share x axis
# fig2, ax2 = plt.subplots()

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

# We will have to use set_title or set_x/ylabel here.
ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
# ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()

```

```python
# Using 2 different figures.

fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

# We will have to use set_title or set_x/ylabel here.
ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()

# fig1.savefig('fig1.png')
# fig2.savefig('fig2.png')
```
