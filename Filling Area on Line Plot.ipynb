```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
# We will use fills in this notebook on the line chart, it can give some professional look
# to our plot and also some useful insights for ex. we can use conditional fills that shows us what areas
# of our line plot is above or below the certain threshold 

# CSV data to work
data = pd.read_csv('data/FillAreaOnLinePlot.csv')
```

```python
# These are ages and corresponding developers salaries
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
plt.plot(ages, dev_salaries, color = '#444444', linestyle = '--', label = 'All Devs')
plt.plot(ages, py_salaries, color = '#4c4aa1', label = 'Python')
plt.legend()
plt.title('Median Salary by Age (USD)')
plt.xlabel('Ages')
plt.ylabel('Median Salaries')
plt.tight_layout()
plt.show()
```

```python
# Suppose we wanna fill all the area under our python plot so to do this create a fill to plot above the
# legend() method and pass x axis and then pass what we wanna fill between here - py_salaries
# We can also pass additional arguement to fill between python and js, this 2nd arguement is called y2
# which is set to 0

# Plotting Median Salaries for all developers
plt.plot(ages, dev_salaries, color = '#444444', linestyle = '--', label = 'All Devs')
# Plotting Median Salaries for python developers
plt.plot(ages, py_salaries, color = '#4c4aa1', label = 'Python')


plt.legend()

# We can also pass aplha arguement to adjust the transparency
plt.fill_between(ages, py_salaries, alpha = 0.25)

plt.title('Median Salary by Age (USD)')
plt.xlabel('Ages')
plt.ylabel('Median Salaries')

plt.tight_layout()
plt.show()
```

```python
# Now if we use filling with the conditionals we can have some additional insights of the data
# Lets say insted of 0 we wanna fill in between the overall median and our python salaries.
# This shows exactly where our python plot cuts the overall median 

# Plotting Median Salaries for all developers
plt.plot(ages, dev_salaries, color = '#444444', linestyle = '--', label = 'All Devs')
# Plotting Median Salaries for python developers
plt.plot(ages, py_salaries, color = '#4c4aa1', label = 'Python')

overall_median = 57287

# We can also pass aplha arguement to adjust the transparency
plt.fill_between(ages, py_salaries, y2= overall_median, alpha = 0.25)

plt.title('Median Salary by Age (USD)')
plt.xlabel('Ages')
plt.ylabel('Median Salaries')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Now we can also fill some extraa conditionals lets say we wanna fill color set green when py_salaries are
# above the median and red when it is below , we will pass an where arguement
# Interpolate  = True will ensure that certain x-intersections dont get clipped and that all of the regions
# are filled correctly

# Plotting Median Salaries for all developers
plt.plot(ages, dev_salaries, color = '#444444', linestyle = '--', label = 'All Devs')
# Plotting Median Salaries for python developers
plt.plot(ages, py_salaries, color = '#4c4aa1', label = 'Python')

overall_median = 57287

# We can also pass aplha arguement to adjust the transparency


# Now due to conditional it is only filling the area above the overall_median, we havent choosen the colors yet

plt.fill_between(ages, py_salaries, y2= overall_median, 
                 where = (py_salaries > overall_median),
                interpolate = True, alpha = 0.25)

plt.title('Median Salary by Age (USD)')
plt.xlabel('Ages')
plt.ylabel('Median Salaries')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plotting Median Salaries for all developers
plt.plot(ages, dev_salaries, color = '#444444', linestyle = '--', label = 'All Devs')
# Plotting Median Salaries for python developers
plt.plot(ages, py_salaries, color = '#4c4aa1', label = 'Python')

overall_median = 57287

# We can also pass aplha arguement to adjust the transparency
# So to fill the are below the overall_median we have to write another fill_between method
# This will automatically set different colors but we can also pass the colors of our choice

# Above
plt.fill_between(ages, py_salaries, y2= overall_median, 
                 where = (py_salaries > overall_median),
                interpolate = True, alpha = 0.25)

#Below
plt.fill_between(ages, py_salaries, y2= overall_median, 
                 where = (py_salaries <= overall_median),
                interpolate = True, color = 'red', alpha = 0.25)

plt.title('Median Salary by Age (USD)')
plt.xlabel('Ages')
plt.ylabel('Median Salaries')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# We can also fill the color between two different plots 

# Plotting Median Salaries for all developers
plt.plot(ages, dev_salaries, color = '#444444', linestyle = '--', label = 'All Devs')
# Plotting Median Salaries for python developers
plt.plot(ages, py_salaries, color = '#4c4aa1', label = 'Python')

overall_median = 57287


# It will be a good idea to label these filled sections , pass label to fill between method
# Above
plt.fill_between(ages, py_salaries, y2= dev_salaries, 
                 where = (py_salaries > dev_salaries),
                interpolate = True, alpha = 0.25, label = 'Above Average')

#Below
plt.fill_between(ages, py_salaries, y2= dev_salaries, 
                 where = (py_salaries <= dev_salaries),
                interpolate = True, color = 'red', alpha = 0.25, label = 'Below Average')


plt.legend()
plt.title('Median Salary by Age (USD)')
plt.xlabel('Ages')
plt.ylabel('Median Salaries')

plt.tight_layout()
plt.show()
```
