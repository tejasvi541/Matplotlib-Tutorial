```python
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
# Pie charts will show us the contribution of each category to the whole data.
# first we need some values we wanna represent in our Piechart, and we can think these as the slices of the pie
slices = [60,40]

plt.pie(slices)

plt.title("My Awesome Pie Chart")
plt.tight_layout
plt.show()
```

```python
# Now lets also add some labels for our chart, We can pass labels as the list and they will correspond
# to the slices we wanted them to be labelled for 
# And its not important for the values to be total of 100 as the pyplot will proportionate them on its
# own to fit that in the pie chart.

slices = [120,80]
labels = ['Sixty', 'Forty']

plt.pie(slices, labels= labels)

plt.title("My Awesome Pie Chart")
plt.tight_layout
plt.show()
```

```python
# Now to seperate the two colors of the pie chart popping against each other we can set some
# wedging properties of it in the label.


slices = [120,80]
labels = ['Sixty', 'Forty']

plt.pie(slices, labels= labels, wedgeprops = {'edgecolor':'black'})
# To get to know abaout all the wedge props attributes check out the matplotlib documentation

plt.title("My Awesome Pie Chart")
plt.tight_layout
plt.show()
```

```python
# Now to change the color of the slices we can do the same we didi with the labels, to pass the 
# colors in the list and the color will get implements corresponding to the slices we wanted the color for.

slices = [120,80,30,20]
labels = ['Sixty', 'Forty', 'Thirty', 'Twenty']
colors = ['blue', 'red', 'yellow', 'green']
# Now we can pass this colors list to out plt method, right after the labels
# We can also use Hex Values 
plt.pie(slices, labels= labels,colors = colors, wedgeprops = {'edgecolor':'black'})


plt.title("My Awesome Pie Chart")
plt.tight_layout
plt.show()
```

```python
# Now using some real world data to plot our pychart, this is the 2019 stackoverflow survey data
# this represents how many respondents have worked with certain languages
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']


plt.pie(slices, labels= labels, wedgeprops = {'edgecolor':'black'})


plt.title("My Awesome Pie Chart")
plt.tight_layout
plt.show()
```

```python
# PieCharts are good for data having 5 entries , more than 5 it just can create a mess, so let
# us manipulate our data to just show 5 entries
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

plt.pie(slices, labels= labels, wedgeprops = {'edgecolor':'black'})


plt.title("My Awesome Pie Chart")
plt.tight_layout
plt.show()
```

```python
# Lots of time you can see a pie chart that have emphasis on one piece of chart, suppose we are preparing 
# a talk on python and wanna emphasise on that so to do this we can pass a explode arguement, this 
# will be a list of values that will offset the slices.
# epl as it isode list will contain floate values and they signifies that how much we wanna emphasize on a data
# So a zero will keep the slice and any number besides 0 will represent fraction of radius we would like to
# explode value means how far it will be, 0.1 is 10%, now we wanna explode python that is the 4th item of the list.

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0]

plt.pie(slices, labels= labels,explode = explode, wedgeprops = {'edgecolor':'black'})


plt.title("My Awesome Pie Chart")
plt.tight_layout
plt.show()
```

```python
# We can add shadow to our chart by passing shadow =True
# We can also choose the starting value , it will set the initial rotation of the chart

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0]

# And to get the percentage of chart each entry is taking we can pass autopct(auto percent) = we will pass
# a format string '%1.1f%%'
plt.pie(slices, labels= labels,explode = explode,shadow = True, startangle =90, autopct = '%1.1f%%'
                                , wedgeprops = {'edgecolor':'black'})

# Now Chart of JS starts at 90 degrees and rotate anti clockwise


plt.title("My Awesome Pie Chart")
plt.tight_layout
plt.show()
```
