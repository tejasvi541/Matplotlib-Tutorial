```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates

plt.style.use('seaborn')
```

```python
# Here we have 7  days back to back and below there is y variable for y axis, lets plot.

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y)

plt.tight_layout()

plt.show()
```

```python
# To get line issted of marker you can explictly mention that 
# You can also turn of the markers by setting marker = None
plt.plot_date(dates, y, linestyle = 'solid')

plt.tight_layout()

plt.show()
```

```python
# Formatting our plot
# running auto format x date method, this will rotate our dates , so they fit a bit nicer and change 
# their alignments and things like that.

# to get a current figure from pyplot we can say plt.gcf and then to run format method we can run this on this
# figure

plt.plot_date(dates, y, linestyle = 'solid')
plt.gcf().autofmt_xdate()
plt.tight_layout()

plt.show()
```

```python
# Lets see how we can change the format of our dates , we want to start with the name of month then day
# and then year , so to do this we will have to run some date time formatting.

plt.plot_date(dates, y, linestyle = 'solid')
plt.gcf().autofmt_xdate() # Here we grab the figure to format figure

date_format = mpl_dates.DateFormatter('%b, %d, %Y') # for this we will grab axis to change format of code
plt.gca().xaxis.set_major_formatter(date_format) # gca means get current axis

plt.tight_layout()

plt.show()
```

### Using Real world data in our csv file

```python
# This is the actual Data of Bitcoin's values

data = pd.read_csv('data/TimeSeriesData.csv')
price_data = data['Date']
price_close = data['Close']
```

```python
# Here this is not plotting our x-axis as our datas but insted this is plotting this out as strings

plt.plot_date(price_data, price_close, linestyle = 'solid')
plt.gcf().autofmt_xdate()

plt.title("Bitcoin Prices")
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()
plt.show()
```

```python
# Plotting x-axis as dates. We will use some pandas methods and after that we will sort them as per dates
# After that if we add some past data to our data set this plot method will plot that accurately.


data['Date'] = pd.to_datetime(data['Date']) # Converting strings to date time object.
data.sort_values('Date', inplace = True) # Sorting Dates

plt.plot_date(price_data, price_close, linestyle = 'solid')
plt.gcf().autofmt_xdate()

plt.title("Bitcoin Prices")
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()
plt.show()
```
