```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Scatter plot will help us to get relate or differentiate 2 different sets of data
plt.style.use('seaborn')

# This is some x, y data between 1-10 to plot
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 9, 1]

# This data shows no corelation here. But creating a scatter plot is as simple as that.
plt.scatter(x, y)

plt.tight_layout()
plt.show()
```

```python
# Now let us change some of the styles of this plot, this time size of dots is small so we will set
# s attribute in plt method 1
# We can also change the color and marker style of the plot , so to change the color we can simply pass
# the c = 'green' and marker style for marker = 'x'

# This is some x, y data between 1-10 to plot
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 9, 1]

# This data shows no corelation here. But creating a scatter plot is as simple as that.
plt.scatter(x, y, s = 100, color = 'green', marker = 'X')

plt.tight_layout()
plt.show()
```

```python
# It is nice to give some edges to the circle and also set the alpha(for tranparency) so we can see
# through them 
# using edgecolor = 'black' and width of the edgecolor using linewidth = 1
# using alpha = 0.75

# This is some x, y data between 1-10 to plot
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 9, 1]

# This data shows no corelation here. But creating a scatter plot is as simple as that.
plt.scatter(x, y, s = 100, color = 'green', edgecolor = 'black', linewidth = 1, alpha = 0.75)

plt.tight_layout()
plt.show()
```

 Now we will see how the color and sizes can actually be on a per mark basis rather than applying 
 them to all of the marks , so hav ing the ability to get multiple color and sizes actually allow us to 
 add additional data set into our plot , now in our current data plot we wanna add some extra information
 for example assume our current data is survey data of bunch of people and we wanna breakdown the data to
 something more specific , so assume we made these people to write something from 1-10 and we somehow want to
 plot their rating as well , to do so we can simply assign different numbers to the different possiblities and then those will give you different colors on your scatter plot as long as you pass that in your method

```python
# So the numbers in the colors list these are no. bw 1-10 , so each of these values will corresponds to our
# data point in our x and y data sets so now if we pass this in our scatter method as the color arguement
# 

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

# This is some x, y data between 1-10 to plot
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 9, 1]

# Now we can also change the color of these by using builtin color maps jus like marker symbols

plt.scatter(x, y, s = 100, c = colors, cmap = 'Greens', edgecolor = 'black', linewidth = 1, alpha = 0.75)
# We will get different shades of green as per intensity
plt.tight_layout()
plt.show()
```

```python
# So you must wanna add the cmap labels so people would know what the color is meant for, to do that we can
# add a colorbar legend 


# This is some x, y data between 1-10 to plot
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 9, 1]

# Now we can also change the color of these by using builtin color maps jus like marker symbols
plt.scatter(x, y, s = 100, c = colors, cmap = 'Greens', edgecolor = 'black', linewidth = 1, alpha = 0.75)

cbar = plt.colorbar()
cbar.set_label('Satisfaction')
plt.tight_layout()
plt.show()
```

```python
# We can also change the size of pur data points as well, so just like color it will add some more ways to
# explain our data so to do this we will define a sizes list


sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]
# This is the list that corresponds to the x and y data set.
# This is some x, y data between 1-10 to plot
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 9, 1]

# Now we can also change the color of these by using builtin color maps jus like marker symbols
plt.scatter(x, y, s = sizes, 
            c = colors, cmap = 'Greens', edgecolor = 'black', linewidth = 1, alpha = 0.75)


cbar = plt.colorbar()
cbar.set_label('Satisfaction')
plt.tight_layout()
plt.show()
```

#### Plotting Real Worl Data Using CSV file Which is some day's you trending page data

```python
data = pd.read_csv('data/ScatterPlotData.csv')
```

```python
# Scatter plot of their total views and total likes and ratio of likes/dislikes as well


view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, edgecolor = 'black', linewidth = 1, alpha = 0.75)

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total likes')

cbar = plt.colorbar()
plt.tight_layout()
plt.show()
```

```python
# We can also use log scale with scatter plot.
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, edgecolor = 'black', linewidth = 1, alpha = 0.75)

plt.xscale('log')
plt.yscale('log')

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total likes')

cbar = plt.colorbar()
plt.tight_layout()
plt.show()
```

```python
# Lets also use the ratio it would be good metric for colors of our points we can also use that for sizes

view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes,c = ratio, cmap = 'summer', edgecolor = 'black', linewidth = 1, alpha = 0.75)

plt.xscale('log')
plt.yscale('log')

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total likes')

cbar = plt.colorbar()
cbar.set_label('Like Dislike Ratio')
plt.tight_layout()
plt.show()
```
